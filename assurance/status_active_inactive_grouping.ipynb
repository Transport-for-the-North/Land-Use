```python
# Check children do not have SOC
status_children = status[(status['Age (D) (5 categories) Code']==1)&(status['Economic activity status (7 categories) Code']>0)].Observation.sum()
print(status_children)
```

```python
# under 75s
status_74 = status[status['Age (D) (5 categories) Code']<5]
```

```python
status_74.Observation.sum()
```

```python
# 75 and over
status_75 = status[(status['Age (D) (5 categories) Code']==5)]
```

```python
status_75.Observation.sum()
```

```python
status_74.Observation.sum()+status_75.Observation.sum()
```

```python
# Remove age bands
```

```python
status_74_no_age = status_74.drop(status_74.columns[[2, 3]], axis=1)
```

```python
status_75_no_age = status_75.drop(status_75.columns[[2, 3]], axis=1)
```

```python
status_74_no_age.Observation.sum()+status_75_no_age.Observation.sum()
```

```python
# Aggregate children and active population at zone and status category with no age
```

```python
status_active_group = status_74_no_age.groupby(['Lower layer Super Output Areas Code',
                       'Lower layer Super Output Areas',
                       'Economic activity status (7 categories) Code',
                       'Economic activity status (7 categories)']).sum().reset_index()
```

```python
status_active_group
```

```python
status_active_group.Observation.sum()
```

```python
# Aggregate over 75 population and replace status with Economically inactive 6
```

```python
status_75_no_age
```

```python
status_inactive_group = status_75_no_age.groupby(['Lower layer Super Output Areas Code',
                       'Lower layer Super Output Areas']).sum().reset_index()
```

```python
status_inactive_group
```

```python
status_inactive_group.Observation.sum()
```

```python
status_inactive_group['Economic activity status (7 categories) Code'] = 5
```

```python
status_inactive_group['Economic activity status (7 categories)'] = 'Economically inactive (excluding full-time students)'
```

```python
status_inactive_group
```

```python
# merge children+active and inactive together
```

```python
status_group = pd.concat([status_active_group,status_inactive_group], axis=0)
```

```python
status_group
```

```python
status_group.Observation.sum()
```

```python
status_group = status_group.groupby(
    ['Lower layer Super Output Areas Code', 'Lower layer Super Output Areas', 'Economic activity status (7 categories) Code', 'Economic activity status (7 categories)'], as_index=False
).agg({'Observation': 'sum'})
```

```python
status_group.Observation.sum()
```
