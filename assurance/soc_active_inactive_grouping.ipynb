```python
soc_74 = soc[soc['Age (D) (5 categories) Code']<5]
```

```python
soc_74.Observation.sum()
```

```python
soc_75 = soc[(soc['Age (D) (5 categories) Code']==5)]
```

```python
soc_75.Observation.sum()
```

```python
soc_74.Observation.sum()+soc_75.Observation.sum()
```

```python
# Remove age bands
```

```python
soc_74_no_age = soc_74.drop(soc_74.columns[[2, 3]], axis=1)
```

```python
soc_75_no_age = soc_75.drop(soc_75.columns[[2, 3]], axis=1)
```

```python
soc_74_no_age.Observation.sum()+soc_75_no_age.Observation.sum()
```

```python
# Aggregate children and active population at zone and soc category with no age
```

```python
soc_active_group = soc_74_no_age.groupby(['Lower layer Super Output Areas Code',
                       'Lower layer Super Output Areas',
                       'Occupation (current) (10 categories) Code',
                       'Occupation (current) (10 categories)']).sum().reset_index()
```

```python
soc_active_group
```

```python
soc_active_group.Observation.sum()
```

```python
# Aggregate over 75 population and replace soc with does not apply
```

```python
soc_75_no_age
```

```python
soc_inactive_group = soc_75_no_age.groupby(['Lower layer Super Output Areas Code',
                       'Lower layer Super Output Areas']).sum().reset_index()
```

```python
soc_inactive_group
```

```python
soc_inactive_group.Observation.sum()
```

```python
soc_inactive_group['Occupation (current) (10 categories) Code'] = -8
```

```python
soc_inactive_group['Occupation (current) (10 categories)'] = 'Does not apply'
```

```python
soc_inactive_group
```

```python
# merge children+active and inactive together
```

```python
soc_group = pd.concat([soc_active_group,soc_inactive_group], axis=0)
```

```python
soc_group
```

```python
soc_group.Observation.sum()
```
