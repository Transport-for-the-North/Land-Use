Define path to the collapsed and recoded NTS data which is the output of `reformat_norcom_data.py`

```python
nts_path = Path(r'I:\NorMITs NorCOM\Import\NTS\nts_hh_data_v2.csv')
```

Define path to store any output analysis

```python
output_path = Path(r'I:\NorMITs NorCOM\AECOM working\nts_analysis')
```

Check the sample sizes of the number of households in the input NTS data that relate to each variable category

```python
# read in nts data
nts_data = pd.read_csv(nts_path)
```

```python
nts_data
```

```python
# define columns of variables that are used in the logit models
variable_cols = ['walkrail_b01id', 'descta_b01id', 'hh_income_banded_v3', 'hh_children_seg', 'hh_adults_seg', 'ns', 'tfn_at']
dependent_cols = ['0v1_car', '1v2+_car']
```

```python
# create list of output dfs to summarise
output = []
```

```python
# get sample size summary for each variable
for variable, dependent in product(variable_cols, dependent_cols):
    df = nts_data.groupby([variable, dependent]).agg({'psuid': 'count'}).reset_index()
    df['variable'] = variable
    df['model'] = dependent
    df = df.rename(columns={variable: 'variable_category', dependent: 'model_category', 'psuid': 'sample'})
    output.append(df)
```

```python
output = pd.concat(output)
```

```python
output.to_csv(output_path / 'nts_sample.csv', index=False)
```
